```python
#8. checking missing values 
a = df.isna().sum()
a
```

```python
#9. heatmap of missing values
plt.figure(figsize=(15,4))
sns.heatmap(df.isnull(),cmap='mako')
```

```python
#10. Countplot of Categorical columns (except name and location)
cat_cols = ['Fuel_Type','Transmission','Owner_Type']
for i in cat_cols:
    plt.figure(figsize=(10,4))
    sns.countplot(df[i])
```

```python
#11. which car has the highest mileage?

Correct_Mileage= []

for i in df.Mileage:
    if str(i).endswith('km/kg'):
        i = i[:-6]
        i = float(i)*1.40
        Correct_Mileage.append(float(i))
    elif str(i).endswith('kmpl'):
        i = i[:-6]
        print(i)
        Correct_Mileage.append(float(i))
    elif str(i) == 'nan':
        i = 'nan'
        Correct_Mileage.append(float(i))
        
        
df['Mileage']= Correct_Mileage
```

```python
#q11.
df.groupby('Name')['Mileage'].max().sort_values(ascending=False).head(1)
```

```python
#12. how many cars whose second hand milage is more than 1st hand mileage  
len(df[(df['Owner_Type'] == 'First') < (df['Owner_Type'] == 'Second')])
```

```python
#13. oldest car in each city
df.groupby(['Location'])['Year','Name'].min()
```

```python
#14. car with highest power in each city 
df.groupby(['Location'])['Power','Name'].max()
```
